```python
#passo a passo
#passo1 Entrar na internet(abrir navegador)

import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

navegador = webdriver.Chrome()
# abrir base de dados


tabela = pd.read_excel("commodities.xlsx")
#import unicodedata


#passo3: Para cada produto da nossa base
#passo4: Atualizar o preço na base de dados
for linha in tabela.index:
    
    
    produto = tabela.loc[linha, "Produto"]
    
    # entrar no site melhor cambio https://www.melhorcambio.com/milho-hoje(formatar caracteres para busca)
    link = f"https://www.melhorcambio.com/{produto}-hoje"
    #link = unicodedata.normalize('NFKD', link).encode('ascii', 'ignore')
    link = link.replace("ó", "o").replace("é", "e").replace("ç", "c").replace("ú", "u").replace("á", "a").replace("ã", "a")
    print(link)
    navegador.get(link)
    
    
    #p3gar cotação e formatando o valor (padrao americano)
    valor_atual = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    valor_atual = valor_atual.replace(".", "").replace(",", ".")
    valor_atual = float(valor_atual)
    print(valor_atual)
    
    #na tabela preço atual atualizar o valor
    # ''  ''  'linha, 'Coluna'
    tabela.loc[linha,"Preço Atual"] = valor_atual

#
#passo6: Decidir quais produtos a gente vai comprar 
#passo7: Exportar a base de dados atualizada
```

```python
#passo2: IMportar Base de dados

display(tabela)

tabela["Comprar"] = tabela["Preço Atual"] < tabela["Preço Ideal"]

display(tabela)

```

```python
navegador.quit()

tabela.to_excel("commodities_atualizado.xlsx")
```
